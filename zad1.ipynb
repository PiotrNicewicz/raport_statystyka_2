<h3>Zad 1</h3>
<p>Z populacji generalnej o rozkładzie normalnym N(µ, σ = 0.2) pobrano próbę (dane dołączone do
listy). Dla α = 0.05 zweryfikuj hipotezy:</br>
▶ µ != 1.5,</br>
▶ µ > 1.5,</br>
▶ µ < 1.5.</br>
Narysuj odpowiednie obszary kytyczne i wyznacz p-wartości dla każdej z powyższych hipotez. Odpowiedz na pytanie co stanie się kiedy zwiększymy bądź zmniejszymy poziom ufności.
</p>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

with open('dane1.txt', 'r', encoding='utf-8') as plik:
    dane = [float(linia) for linia in plik]

sigma = 0.2
mu0 = 1.5
alpha = 0.05

n = len(dane)
x_sr = np.mean(dane)

Z = (x_sr - mu0)/(sigma/np.sqrt(n))

print("Statystyka Z =", round(Z, 5))

#1. H1: mi != 1.5

z_kryt = norm.ppf(1 - alpha / 2)
p_value = 2 * (1 - norm.cdf(abs(Z)))
print("H1: mi != 1.5")
print("Wartość krytyczna:", z_kryt)
print("p-value =", p_value)

if abs(Z) > z_kryt:
    print("Odrzucamy H0")
else:
    print("Brak podstaw do odrzucenia H0")

#2. H1: mi > 1.5

z_kryt = norm.ppf(1 - alpha)
p_value = 1 - norm.cdf(Z)

print("H1: mi > 1.5")
print("Wartość krytyczna:", z_kryt)
print("p-value =", p_value)

if Z > z_kryt:
    print("Odrzucamy H0")
else:
    print("Brak podstaw do odrzucenia H0")

# 3. H1: mi < 1.5

z_kryt = norm.ppf(alpha)
p_value = norm.cdf(Z)

print("H1: mi < 1.5")
print("Wartość krytyczna:", z_kryt)
print("p-value =", p_value)

if Z < z_kryt:
    print("Odrzucamy H0")
else:
    print("Brak podstaw do odrzucenia H0")

```
